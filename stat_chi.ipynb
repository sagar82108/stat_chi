```python
import scipy.stats as stats
import numpy as np

# Observed frequencies of each color in the sample
observed_counts = np.array([30, 25, 22, 11, 12, 20])  # Replace with your actual counts

# Expected percentages based on the expected distribution
expected_percentages = np.array([20, 20, 20, 10, 10, 20])  # Replace with your expected percentages

# Sample size
sample_size = sum(observed_counts)

# Calculate expected counts based on percentages
expected_counts = (expected_percentages / 100) * sample_size

# Perform the chi-square goodness of fit test
chi_squared, p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Define the significance level
alpha = 0.05

# Calculate degrees of freedom
degrees_of_freedom = len(observed_counts) - 1

# Calculate the critical chi-square value
critical_chi_squared = stats.chi2.ppf(1 - alpha, degrees_of_freedom)

# Compare the calculated chi-square statistic to the critical value
if chi_squared > critical_chi_squared:
    print("Reject the null hypothesis. The distribution does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution matches the expected distribution.")

# Print the chi-square statistic and p-value
print(f"Chi-square Statistic: {chi_squared}")
print(f"P-value: {p_value}")
```

```python
import scipy.stats as stats
import numpy as np

# Define the observed counts for the two groups
group1 = np.array([20, 10, 15])
group2 = np.array([15, 25, 20])

# Perform the chi-square test of independence
chi_squared, p_value, _, _ = stats.chi2_contingency([group1, group2])

# Print the chi-square statistic and p-value
print(f"Chi-square Statistic: {chi_squared}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant association between the two groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between the two groups.")
```

```python
import scipy.stats as stats

# Sample size and number of individuals who smoke in the sample
n = 500
x = 60

# Calculate the sample proportion (p-hat)
p_hat = x / n

# Calculate the standard error
standard_error = ((p_hat * (1 - p_hat)) / n) ** 0.5

# Set the confidence level
confidence_level = 0.95

# Calculate the critical values for a two-tailed test
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_score * standard_error

# Calculate the confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

print(f"95% Confidence Interval for Proportion: ({lower_bound:.4f}, {upper_bound:.4f})")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Define the degrees of freedom
df = 10

# Generate x values for the chi-square distribution
x = np.linspace(0, 30, 1000)  # Adjust the range as needed

# Calculate the chi-square probability density function (PDF) for the given degrees of freedom
pdf = stats.chi2.pdf(x, df)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the chi-square PDF
ax.plot(x, pdf, label=f'Chi-square with {df} degrees of freedom')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
pdf_fill = stats.chi2.pdf(x_fill, df)
ax.fill_between(x_fill, pdf_fill, color='gray', alpha=0.5, label='Chi-square < 15')

# Label the axes
ax.set_xlabel('Chi-square Statistic')
ax.set_ylabel('Probability Density')
ax.set_title(f'Chi-square Distribution with {df} Degrees of Freedom')
ax.legend()

# Show the plot
plt.show()
```

```python
import scipy.stats as stats

# Step 1: Define your observed and expected frequencies
observed_frequencies = [45, 55]  # 45 tails and 55 heads
expected_frequencies = [50, 50]  # Expected frequencies for a fair coin

# Step 2: Set the significance level (alpha)
alpha = 0.05

# Step 3: Calculate the chi-square test statistic and p-value
chi2, p = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Step 4: Determine the degrees of freedom
degrees_of_freedom = len(observed_frequencies) - 1

# Step 5: Find the critical value (optional)
critical_value = stats.chi2.ppf(1 - alpha, degrees_of_freedom)

# Step 6: Print the results
print(f"Chi-square test statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {degrees_of_freedom}")
print(f"Critical value: {critical_value}")

# Step 7: Make a decision based on the p-value
if p <= alpha:
    print("Reject the null hypothesis: The coin is biased towards tails.")
else:
    print("Fail to reject the null hypothesis: The coin is likely fair.")
```

```python
import scipy.stats as stats

# Given data
sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean = 70
significance_level = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1

# Find the critical t-values for a two-tailed test
alpha = significance_level / 2  # Since it's a two-tailed test
t_critical_left = stats.t.ppf(alpha, degrees_of_freedom)
t_critical_right = stats.t.ppf(1 - alpha, degrees_of_freedom)

# Print the results
print(f"Calculated t-statistic: {t_statistic}")
print(f"Critical t-values (left and right): {t_critical_left}, {t_critical_right}")

# Perform the hypothesis test
if t_statistic < t_critical_left or t_statistic > t_critical_right:
    print("Reject the null hypothesis: Population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: Population mean is not significantly different from 70.")
```
